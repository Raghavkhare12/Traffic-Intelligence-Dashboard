```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/events_features.csv"
)

df.head()
```

```python
pd.crosstab(
    df["event_cause"],
    df["risk_level"]
)
```

```python
resource_map = {
    "Critical": 20,
    "High": 10,
    "Medium": 5,
    "Low": 2
}

df["officer_score"] = (
    df["risk_level"]
    .map(resource_map)
)
```

```python
resource_table = (
    df.groupby(
        "event_cause"
    )["officer_score"]
    .mean()
    .round()
    .reset_index()
)

resource_table
```

```python
resource_table["officers"] = (
    resource_table["officer_score"]
)

resource_table["barricades"] = (
    resource_table["officers"] // 2
)

resource_table["diversions"] = (
    resource_table["officers"] // 5
)

resource_table
```

```python
resource_table.sort_values(
    "officers",
    ascending=False
)
```

```python
resource_table.to_csv(
    "../data/resource_plan.csv",
    index=False
)
```
