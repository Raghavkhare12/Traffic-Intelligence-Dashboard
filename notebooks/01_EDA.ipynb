```python
datetime_cols = [
    "start_datetime",
    "end_datetime",
    "created_date",
    "closed_datetime",
    "resolved_datetime"
]

for col in datetime_cols:
    df[col] = pd.to_datetime(
        df[col],
        errors="coerce"
    )
```

```python
df[datetime_cols].dtypes
```

```python
df["hour"] = df["start_datetime"].dt.hour

df["day_of_week"] = df["start_datetime"].dt.day_name()

df["month"] = df["start_datetime"].dt.month

df["year"] = df["start_datetime"].dt.year
```

```python
plt.figure(figsize=(12,5))

df["hour"].value_counts().sort_index().plot()

plt.title("Events by Hour")

plt.xlabel("Hour")

plt.ylabel("Count")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10,5))

order = [
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"
]

df["day_of_week"].value_counts().reindex(order).plot(
    kind="bar"
)

plt.title("Events by Day")

plt.show()
```

```python
df["duration_hours"] = (
    df["end_datetime"] -
    df["start_datetime"]
).dt.total_seconds() / 3600
```

```python
df["duration_hours"].describe()
```

```python
cause_weights = {
    "vehicle_breakdown": 2,
    "pot_holes": 3,
    "water_logging": 4,
    "tree_fall": 5,
    "road_conditions": 5,
    "accident": 7,
    "congestion": 8,
    "construction": 9,
    "public_event": 9,
    "procession": 10,
    "vip_movement": 10,
    "protest": 10
}
```

```python
df["cause_score"] = (
    df["event_cause"]
    .map(cause_weights)
    .fillna(3)
)
```

```python
df["closure_score"] = (
    df["requires_road_closure"]
    .astype(int)
    * 5
)
```

```python
priority_weights = {
    "Low": 1,
    "High": 5
}

df["priority_score"] = (
    df["priority"]
    .map(priority_weights)
)
```

```python
df["impact_score"] = (
    df["cause_score"] +
    df["closure_score"] +
    df["priority_score"]
)
```

```python
df["impact_score"].describe()
```

```python
def risk_level(score):

    if score <= 5:
        return "Low"

    elif score <= 10:
        return "Medium"

    elif score <= 15:
        return "High"

    else:
        return "Critical"
```

```python
df["risk_level"] = (
    df["impact_score"]
    .apply(risk_level)
)
```

```python
df["risk_level"].value_counts()
```
